```python
from xmlrpc.client import ServerProxy
import pandas as pd
```

```python
df_data = pd.read_csv("./data/train_A.csv")
df_test = pd.read_csv("./data/test.csv")
```

```python

# Dados de treino
features_train = df_data.drop(columns=['SalePrice']).to_dict(orient='list')

labels_train = df_data['SalePrice'].tolist()

# Dados para previsão
features_test = df_data.drop(columns=['SalePrice']).to_dict(orient='list')
```

```python
server = ServerProxy("http://localhost:8000/")
```

```python
print("Inicializando o modelo ...")
# server.initModel("dummy")
server.initModel("linearRegression")
```

```python
print("Treinando o modelo com os dados de treino...")
server.fit(features_train, labels_train)
```

```python
print("Fazendo previsões com novos dados...")
predictions = server.predict(features_test)
```

```python
# 4. Exibe as previsões
print("Previsões recebidas do servidor:", predictions)
```
